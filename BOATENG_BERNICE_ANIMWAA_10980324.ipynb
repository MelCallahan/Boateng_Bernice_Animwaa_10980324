## Removing null values:

### We must eliminate any null values or duplicates from the dataset. There are several ways to deal with null values, including removing the rows or columns containing null values or filling them with the mean or median of the column. The technique chosen is determined by the data and the analysis being done. With the drop duplicates() function, we can easily get rid of duplicates.
